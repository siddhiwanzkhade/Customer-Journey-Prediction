```python
orders = orders.merge(
    customers[["customer_id", "customer_unique_id"]],
    on="customer_id",
    how="left"
)
```

```python
orders.groupby("customer_unique_id")["order_id"].count().value_counts()
```

```python
#aggregating payments
payments_agg = order_payments.groupby("order_id", as_index=False).agg({
    "payment_value": "sum",
    "payment_installments": "max"
})
```

```python
#merging orders + items
orders_items = orders.merge(order_items, on="order_id", how="left")
```

```python
#add payments 
orders_items_pay = orders_items.merge(payments_agg, on="order_id", how="left")
```

```python
orders_items_pay.head()
orders_items_pay.shape
```

```python
#CREATING USER LEVEL DATASET
user_df = orders_items_pay.groupby("customer_unique_id").agg({
    "order_id": "nunique",        # total orders
    "payment_value": "sum",       # total spend
    "price": "mean",              # avg price
    "order_item_id": "count"      # total items
}).reset_index()
```

```python
#renameing columns
user_df.columns = [
    "customer_unique_id",
    "total_orders",
    "total_spent",
    "avg_price",
    "total_items"
]
```

```python
user_df["avg_order_value"] = user_df["total_spent"] / user_df["total_orders"]
```

```python
#fixing date dtype
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders_items_pay[col] = pd.to_datetime(orders_items_pay[col])
```

```python
#how recently a user bought something
last_purchase = orders_items_pay.groupby("customer_unique_id")["order_purchase_timestamp"].max()

user_df["days_since_last_purchase"] = user_df["customer_unique_id"].map(
    (orders_items_pay["order_purchase_timestamp"].max() - last_purchase).dt.days
)
```

```python
#tenure(how long the user has been active )
first_purchase = orders_items_pay.groupby("customer_unique_id")["order_purchase_timestamp"].min()

user_df["customer_tenure_days"] = user_df["customer_unique_id"].map(
    (orders_items_pay["order_purchase_timestamp"].max() - first_purchase).dt.days
)
```

### CREATING TARGET VARIABLE SO THAT MODEL CAN LEARN PATTERNS


```python
#REFERENCE DATE(WICH ACTS LIKE TODAY)
max_date = orders_items_pay["order_purchase_timestamp"].max()
```

```python
#LAST PURCHASE DATE PER USER
last_purchase = orders_items_pay.groupby("customer_unique_id")[
    "order_purchase_timestamp"
].max().reset_index()
```

```python
#CREATE TARGET (labels users as 1 if they purchased in last 90 days , else 0)
last_purchase["target_90d"] = (
    (max_date - last_purchase["order_purchase_timestamp"]).dt.days <= 90
).astype(int)
```

```python
#merge the target variable into user_df
user_df = user_df.merge(
    last_purchase[["customer_unique_id", "target_90d"]],
    on="customer_unique_id",
    how="left"
)
```

```python
#filling missing na
user_df["target_90d"] = user_df["target_90d"].fillna(0)
```
